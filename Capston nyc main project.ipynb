### CREATING TABLE

```python
# 1st step Create schema and SQL Tables

def create_tables():
    conn = get_db_connection()
    cursor = conn.cursor()

    create_table_query = '''
                            CREATE SCHEMA IF NOT EXISTS  capston;

                            -- Drop tables in reverse order of dependency
                            DROP TABLE IF EXISTS  capston.Capston_fact_df CASCADE;
                            DROP TABLE IF EXISTS  capston.Payment_df CASCADE;
                            DROP TABLE IF EXISTS  capston.Employee_df CASCADE;
                            DROP TABLE IF EXISTS  capston.Agency_df CASCADE;

                            -- Create tables in order of dependency
                            CREATE TABLE IF NOT EXISTS  capston.Agency_df (
                               Agency_ID SERIAL PRIMARY KEY,
                               Agency_Code INTEGER,
                               Agency_Name TEXT,
                               Work_Location TEXT,
                               Agency_start_Date DATE
                            );

                            CREATE TABLE IF NOT EXISTS  capston.Employee_df (
                               Employee_ID SERIAL PRIMARY KEY,
                               First_Name TEXT,
                               Last_Name TEXT,
                               Title_ID INTEGER,
                               Title_Description TEXT,
                               Leave_Status TEXT,
                               Agency_ID INTEGER,
                               FOREIGN KEY(Agency_ID) REFERENCES  capston.Agency_df(Agency_ID)
                            );

                            CREATE TABLE IF NOT EXISTS  capston.Payment_df (
                               Payment_ID SERIAL PRIMARY KEY,
                               Payroll_Number INTEGER,
                               Base_Salary FLOAT,
                               Basic_pay TEXT,
                               OverTime_Pay FLOAT,
                               Other_Pay FLOAT,
                               Gross_Pay FLOAT
                            );

                            CREATE TABLE IF NOT EXISTS  capston.Capston_fact_df (
                               capston_fact_ID SERIAL PRIMARY KEY,
                               Agency_ID INTEGER,
                               Employee_ID INTEGER,
                               Payment_ID INTEGER,
                               Regular_Hours TEXT, 
                               OverTime_Hours TEXT,
                               Fiscal_Year INTEGER,
                               FOREIGN KEY(Agency_ID) REFERENCES  capston.Agency_df(Agency_ID),
                               FOREIGN KEY(Employee_ID) REFERENCES  capston.Employee_df(Employee_ID),
                               FOREIGN KEY(Payment_ID) REFERENCES  capston.Payment_df(Payment_ID)
                            );
                            '''
    cursor.execute(create_table_query)
    conn.commit()
    cursor.close()
    conn.close()
```

```python
import csv

def load_data_from_csv(csv_path):
    conn = get_db_connection()
    cursor = conn.cursor()
    with open(csv_path, 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            try:
              cursor.execute('''
                    INSERT INTO capston.Agency_df (Agency_ID, Agency_Code, Agency_Name, Work_Location, Agency_start_Date)
                    VALUES (%s, %s, %s, %s,%s);
                ''', row)
            except psycopg2.IntegrityError:
                conn.rollback()
            else:    
                conn.commit()
    
    conn.commit()
    cursor.close()
    conn.close()

csv_file_path = r'C:\Users\back2\Desktop\CAPSTONE PROJECT NYC\CAPSTON-NYC-PROJECT\'Agency_df.csv'
load_data_from_csv(csv_file_path)
```
